```python
import os
import pandas as pd
import numpy as np
import pathlib
import xarray as xr
```

#### 1- 读取原始文件

```python
read_file_path: str = r'./../data/NMF_TRN_OSTZSS_CSDT_2023052112_168h_SS_maxSurge.txt'
# 读取含中文表头的 tab
with open(read_file_path, 'rb') as f:
    # pandas.core.frame.DataFrame
    data: pd.core.frame.DataFrame = pd.read_csv(f, encoding='gbk', sep='\s+', header=None,
                                                infer_datetime_format=False)
    print('读取成功')
```

```python
data
```

```python
# 定义经纬度数组
# 注意经纬度网格的 长宽 无问题，但是范围有问题
# 250
lon=np.linspace(105,136.25,250)
# 222
lat=np.linspace(12.125,41.875,220)
```

```python
# ValueError: conflicting sizes for dimension 'lon': length 220 on the data but length 250 on coordinate 'lon'
da = xr.DataArray(data, coords=[lat, lon], dims=['lat', 'lon'])
da
```

```python
lon=np.linspace(105,136.25,250)
lat=np.linspace(12.125,41.875,222)
```

```python
# 需要将lat与lon的ndarray->dataarray
lat_arr=xr.DataArray(lat)
lon_arr=xr.DataArray(lon)
```

```python
type(lat_arr)
```

```python
# lon, lat = np.meshgrid(lon, lat)
coords = {'lon':  lon_arr, 'lat':  lat_arr}
```

直接针对读取后的 data 声明为 DataArray 并声明坐标

```python
# TypeError: unhashable type: 'numpy.ndarray'
# ValueError: inferring DataArray dimensions from dictionary like ``coords`` is no longer supported. Use an explicit list of ``dims`` instead.
# coords 坐标系不允许通过字典声明，改为dims
da=xr.DataArray(data, dims={'lat': lat_arr, 'lon': lon_arr})
```

```python
#ValueError: conflicting sizes for dimension 'dim_0': length 250 on 'lon' and length 222 on 'lat'
da.assign_coords(lon=lon_arr,lat=lat_arr)
```
