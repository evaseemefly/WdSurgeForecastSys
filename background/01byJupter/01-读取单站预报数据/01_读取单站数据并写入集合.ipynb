### 1- 读取单点站预报数据

```python
read_file_path: str = r'./../data/NMF_TRN_OSTZSS_CSDT_2023051612_168h_SS_staSurge.txt'
# 读取含中文表头的 tab
with open(read_file_path, 'rb') as f:
    data = pd.read_table(f, encoding='gbk', sep='\s+',
                         header=0, infer_datetime_format=False)
    print('读取成功')
```
